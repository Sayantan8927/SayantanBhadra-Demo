Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

Data Collection

```python
# loading  the dataset to a Pandas DataFrame
wine_dataset = pd.read_csv('/content/winequality-red.csv')
```

```python
# number of rows & columns in the dataset
wine_dataset.shape
```

```python
# first 5 rows of the dataset
wine_dataset.head()
```

```python
# checking for missing values
wine_dataset.isnull().sum()
```

```python
# statical measures of the datset
wine_dataset.describe()
```

Data Processing

```python
# separate the data and Lable
X = wine_dataset.drop('quality', axis=1)
```

```python
print(X)
```

Lable Binarization

```python
Y = wine_dataset['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
```

```python
print(Y)
```

Train & Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

Model Training

Random Forest Classifier

```python



model =  RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

Model Evalution

Accuracy Score

```python
# accuracy on test data
X_test_predection = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_predection, Y_test)
```

```python
print('Accuracy : ', test_data_accuracy)
```

Bulding a Predictive System

```python
input_data = (7.5, 0.5, 0.36, 6.1, 0.071, 17, 102, 0.9978, 3.35, 0.8, 10.5)

# changing the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the data as we are predicting the label for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# Get feature names from the training data
feature_names = X_train.columns

# Convert input data to a pandas DataFrame with feature names
input_data_df = pd.DataFrame(input_data_reshaped, columns=feature_names)

predection = model.predict(input_data_df)
print(predection)

if (predection[0]==1):
  print('Good Quality Wine')
else:
  print('Bad Quality Wine')

prediction = model.predict(input_data_df)
print(prediction)
```
